## SGD regression with the Boston housing data

```python
from sklearn.datasets import load_boston
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df = load_boston()
```

```python
data1 = df.data
target1 = df.target
```

```python
X_train1,X_test1,y_train1,y_test1 = train_test_split(data1,target1)
```

```python
X_scaler = StandardScaler()
y_scaler = StandardScaler()
```

```python
X_train1 = X_scaler.fit_transform(X_train1)
y_train1 = y_scaler.fit_transform(y_train1)
X_test1 = X_scaler.fit_transform(X_test1)
y_test1 = y_scaler.fit_transform(y_test1)
```

```python
clf3 = SGDRegressor()
scores = cross_val_score(clf3,X_train1,y_train1,cv = 5)
```

```python
print scores.mean(),scores
```

Now test the R score on the test set

```python
clf3.fit_transform(X_train1,y_train1)
y_pred1 = clf3.predict(X_test1)
```

```python
print 'the R score on the test set is {}'.format(clf3.score(X_test1,y_test1))
```

### SGD with l1 loss

```python
clf4 = SGDRegressor(penalty= 'l1',alpha = 0.01)
```

```python
X_train1_new = clf4.fit_transform(X_train1,y_train1)
```

```python
X_train1.shape
```

```python
X_train1_new.shape
```

```python
y_pred2 = clf4.predict(X_test1)
```

```python
clf4.score(X_test1,y_test1)
```

CV to get the best alpha 

```python
parameters = {'alpha':np.logspace(-4,-1,50)}
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
gs = GridSearchCV(clf4,param_grid = parameters, cv = 5,n_jobs = -1)
```

```python
gs.fit(X_train1,y_train1)
```

```python
print gs.best_score_, gs.best_params_
```

```python
X_train1_new2 = gs.transform(X_train1)
```

```python
X_train1_new2.shape
```

Therefore we reach the conclusion that no dimension reduction is optimal for the SG regression

```python
gs.score(X_test1,y_test1)
```

### CV with SGD with elasticNet

```python
clf5 = SGDRegressor(penalty='elasticnet')
```

```python
params = {'l1_ratio':np.linspace(0,1,20),'alpha':np.logspace(-4,-1,50)}
gs2 = GridSearchCV(clf5,param_grid = params,cv = 5,n_jobs = -1)
```

```python
gs2.fit(X_train1,y_train1)
```

```python
gs2.best_params_
```

```python
gs2.best_score_
```

```python
gs2.score(X_test1,y_test1)
```

```python
clf6 = gs2.best_estimator_
```

```python
clf6.score(X_test1,y_test1)
```

```python
gs2.grid_scores_
```

```python
X_train1.shape
```
